9. Simular dos variables aleatorios X1 y X2 que siguen distribuciones de Poisson con parametros μ1 = 1 y μ2 = 2 respectivamente. Calcular la variable Y = X1 + X2. Repetir la simulacion 1000 veces. Graficar la distribucion de la variable Y y comparar con una distribucion de probabilidad adecuada.

```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Defino la variable de Poisson que es la suma, lo hago como funcion para experimentar con los valores
def PoissVar(n=1000, mu1=1, mu2=2):
    #Usamos comando que toma n elementos de una distribucion de poisson con parametro mu
    x1 = np.random.poisson(mu1,n)
    x2 = np.random.poisson(mu2,n)

    y = x1 + x2

    return y

#Ejemplo para 10 simulaciones
PoissVar(1000)
```

```python
fig, ax = plt.subplots(1,1)

#Uso la funcion plt.hist ya que los datos no han sido bineados, aca hace eso y el grafico de barras a la vez.
ax.hist(PoissVar(1000), bins = range(15), density = True, width = 0.8,label= 'Histograma')

#Grafico la distribucion de Poisson
x = np.arange(0,15)
ax.plot(poisson.pmf(x,3), label = 'Poisson')

#Etiquetas de figura
ax.legend()
ax.set_title('Histograma comparado con distribucion de probabilidad de Poisson')
ax.set_xlabel('Valores de Y')
ax.set_ylabel('Probabilidad')
ax.grid(True)
```
